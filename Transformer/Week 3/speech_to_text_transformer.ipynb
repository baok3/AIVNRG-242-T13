```python
from datasets import load_dataset

dataset = load_dataset("mozilla-foundation/common_voice_11_0", "vi")  # Ngôn ngữ Vietnamese
print(dataset)
```

```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration

model_name = "openai/whisper-small"  # Hoặc thử "openai/whisper-large-v2"
processor = WhisperProcessor.from_pretrained(model_name)
model = WhisperForConditionalGeneration.from_pretrained(model_name)
```

```python
import librosa
import numpy as np

def preprocess_function(batch):
    audio = batch["audio"]

    # Resample bằng librosa (tránh lỗi kiểu dữ liệu)
    resampled_waveform = librosa.resample(audio["array"], orig_sr=audio["sampling_rate"], target_sr=16000)

    # Ép kiểu về float32
    resampled_waveform = np.array(resampled_waveform, dtype=np.float32)

    batch["input_features"] = processor(resampled_waveform, sampling_rate=16000).input_features[0]
    batch["labels"] = processor.tokenizer(batch["sentence"]).input_ids
    return batch
```

```python
dataset = dataset.map(preprocess_function, remove_columns=["audio", "sentence"])
```

```python
from transformers import WhisperFeatureExtractor, DataCollatorForSeq2Seq

feature_extractor = WhisperFeatureExtractor.from_pretrained(model_name)

data_collator = DataCollatorForSeq2Seq(
    feature_extractor,  # <== Dùng feature_extractor thay vì processor
    model,
    padding=True
)
```

```python
from transformers import TrainingArguments, Trainer
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
# Thiết lập tham số huấn luyện
training_args = TrainingArguments(
    output_dir="./whisper-finetuned-vi",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=5e-5,
    warmup_steps=500,
    num_train_epochs=3,
    logging_dir="./logs",
    fp16=True if device == "cuda" else False,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    tokenizer=processor.feature_extractor,
    data_collator=data_collator,
)

# Bắt đầu fine-tuning
trainer.train()
```

```python
import torch
import librosa
from transformers import WhisperProcessor, WhisperForConditionalGeneration

def transcribe(audio_path, model_path="/content/whisper-finetuned-vi"):
    # Load mô hình đã fine-tune
    processor = WhisperProcessor.from_pretrained(model_path)
    model = WhisperForConditionalGeneration.from_pretrained(model_path)
    model.eval()

    # Load và chuẩn bị âm thanh
    audio, sr = librosa.load(audio_path, sr=16000)
    input_features = processor(audio, sampling_rate=16000, return_tensors="pt").input_features

    # Dự đoán văn bản
    with torch.no_grad():
        predicted_ids = model.generate(input_features)

    # Giải mã output thành văn bản
    return processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]

# Ví dụ chạy thử
text = transcribe("/content/drive/MyDrive/ghi âm.mp3")
print("Transcription:", text)
```
